# CUDA vs Tensor Cores

```python
mat_sizes = ["128 x 128", "256 x 256", "512 x 512", "1028 x 1028", "2048 x 2048", "4096 x 4096"]
tiled_gflops = [313.195, 1640.19, 1928.46, 2510.58, 2712.97, 2971.74]
tensor_gflops = [509.809, 3273.73, 11399.1, 13933.1, 15124.7, 17406.9]

# Create the bar plot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the bars
bars_2 = ax.bar(mat_sizes[1:], tensor_gflops[1:], color='r', label="Tensor Cores (Naive Version)")
bars_1 = ax.bar(mat_sizes[1:], tiled_gflops[1:], color='g', label="CUDA Cores (Tiled Version)")
# plt.semilogy()
# plt.ylim(1e+2, 5e+4)

# Add labels and title
ax.set_xlabel('Matrices of size (n x n)')
ax.set_ylabel('GFLOPS')
plt.legend()

# Add value labels on top of each bar
for (i, bar) in enumerate(bars_1):
    height = bar.get_height()
    if i == 0 or i == 1:
        ax.text(bar.get_x() + bar.get_width()/2., height+100,
                f'{height}',
                ha='center', va='bottom')
    else:
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{height}',
                ha='center', va='bottom')
for bar in bars_2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{height}',
            ha='center', va='bottom')


plt.show()
```
